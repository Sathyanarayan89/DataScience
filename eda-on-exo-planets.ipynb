### Data Import

```python
import pandas as pd;

exo_planets = pd.read_csv('../input/nasa-exoplanetary-system/exoplanets.csv');
exo_planets.info()
```

#### lets display first 20 null columns

```python
print(exo_planets.isnull().sum().sort_values(ascending = False).head(20));
```

#### Lets visualize the missing data in each columns

- The bar chart from missingo library gives us an idea of how many missing data are there.
- The white regions are NaN while fully black regions means no Nan.

```python
import missingno as msno
msno.matrix(exo_planets);
```

```python
filtered_df = exo_planets.dropna(axis=1);
msno.matrix(filtered_df);
filtered_df.info();
```

- There are only 21 columns where data is available for all exo-planets discovered.

### Analysis on Discovery facility

```python
print("No. of unique astro observatory facility :", len(pd.unique(exo_planets['disc_facility'])));
```

- there seems to be 65 unique observatory. Lets see the counts of each observatory.

```python
# count no. of occurances of each observatories.
import matplotlib.pyplot as plt

df_facility = exo_planets.disc_facility.value_counts();
fig1, ax = plt.subplots(1,1,figsize=(16, 5),dpi=100)
df_facility.plot.bar(log=True,ax=ax)
```

##### It seems most of the discovery is made by Kepler observatory. Kepler leads others by a log scale!
